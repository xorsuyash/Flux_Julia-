```julia
using Flux, Statistics ,ProgressMeter 
```

```julia
noisy=rand(Float32,2,1000)
```

```julia
truth=[xor(col[1]>0.5,col[2]>0.5) for col in eachcol(noisy)]
```

```julia
model=Chain(Dense(2 =>3,tanh),
    BatchNorm(3),
Dense(3 => 2),softmax)
```

```julia
out1=model(noisy)
```

```julia
target=Flux.onehotbatch(truth,[true,false])
```

```julia
target
```

```julia
loader=Flux.DataLoader((noisy,target),batchsize=64,shuffle=true);
```

```julia
optim = Flux.setup(Flux.Adam(0.01),model)
```

```julia
losses=[]
```

```julia
@showprogress for epoch in 1:1_1000
    for (x,y) in loader
        loss,grad=Flux.withgradient(model) do m
            y_hat=m(x)
            Flux.crossentropy(y_hat,y)
        end 
        Flux.update!(optim, model, grad[1])
        push!(losses ,loss)
    end 
end
```

```julia
optim
```

```julia
out2=model(noisy)
```

```julia
mean((out2[1,:].>0.5).==truth)
```

```julia
plot(losses; xaxis=(:log10, "iteration"),
    yaxis="loss", label="per batch")
n = length(loader)
plot!(n:n:length(losses), mean.(Iterators.partition(losses, n)),
    label="epoch mean", dpi=200)
```
